```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/amy/projects/Tigo/data/cleaned_data.csv')
```

```python
df
```

```python
# Convert to datetime
df['FECHA_EMISION'] = pd.to_datetime(df['FECHA_EMISION'])

# Create 'cohort_month' column
df['cohort_month'] = df['FECHA_EMISION'].dt.to_period('M').dt.to_timestamp('D')
```

```python
df
```

```python
# Create the pivot table
pivot_table_expected = df.pivot_table(
    index=None,          # No row index, since we only want the column sums
    columns='cohort_month',
    values='MONTO_FACTURA',
    aggfunc='sum'
)
```

```python
pivot_table_expected
```

```python
pivot_table_expected.to_csv('/home/amy/projects/Tigo/output/expected.csv')
```
